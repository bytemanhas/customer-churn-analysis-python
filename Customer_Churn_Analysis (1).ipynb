```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

df = pd.read_csv("customer_churn_dataset.csv")
df.head()
```

Business Problem:

Customer churn is a major issue for companies. The goal is to analyze customer data and identify patterns and factors that lead to customer churn.

This makes your project look professional.

```python
#Customer Churn Count Analysis
df['Churn'].value_counts()
```

```python
df['Churn'].value_counts().plot(kind='bar')
plt.title("Customer Churn Count")
plt.show()
```

```python
#Average Monthly Charges by Churn
df.groupby('Churn')['MonthlyCharges'].mean()
```

```python
sns.barplot(x='Churn', y='MonthlyCharges', data=df)
plt.show()
```

```python
#Churn by Contract Type
df.groupby('Contract')['Churn'].value_counts()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x='Churn', y='Tenure', data=df)
plt.title("Tenure vs Churn")
plt.show()
```

```python
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Churn by Contract Type")
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

```python
df.groupby('Contract')['Churn'].value_counts()
```

Key Insights:

Example (based on your analysis):

• Customers with month-to-month contracts have higher churn

• Customers with higher monthly charges tend to churn more

• Customers with lower tenure churn more

• Long-term contract customers churn less
